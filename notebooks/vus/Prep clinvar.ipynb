Prepare a clinvar.txt file for VUS cooccurrence analysis

```python
import re
import json
file1 = open('/Users/forei/Downloads/clinvar_result.txt', 'r')


count = 0

header = file1.readline()
fields = header.split('\t')
#print(fields)
clinsigs = {
   "Benign": "edit",
   "Conflicting interpretations of pathogenicity": "edit",
   "Uncertain significance": "-",
   "Likely benign": "edit",
   "Pathogenic/Likely pathogenic": "edit",
   "Likely pathogenic": "edit",
   "Pathogenic": "edit",
   "not provided": "edit",
   "Benign/Likely benign": "edit"
}


outputsigs = {
   #"Benign": "Benign",
   "Uncertain significance": "-",
   "Pathogenic/Likely pathogenic": "Pathogenic",
   "Pathogenic": "Pathogenic",
}

outputDict = {"Pathogenic":[], "-":[]}

while True:
    count += 1
 
    # Get next line from file
    line = file1.readline()
 

    if not line:
        break
        
    fields = line.split('\t')
    #Name, Gene, px, cx, clinsig, rev_status, accession, GRCh37Chromosome, GRCh37Location, 
    #GRCh38Chromosome, GRCh38Location, VariationID, AlleleIDs, dbSNP, CanonicalSPDI = 
    nm, cs, GRCh37Chromosome, GRCh37Location, GRCh38Location =  [fields[i] for i in (0, 4, 7, 8, 10)]
    locs = GRCh37Location.split('-')
    p = re.compile('.*([TCGA]>[TCGA]*)$')
    if len(locs) <2:
        #print (locs[0])
        clinsig = cs.split('(')[0]
        if clinsig not in clinsigs:
            clinsigs[clinsig] = 'edit'
        #print(nm)
        m = p.match(nm)
        if m:
            #print("match {}".format(m))
            #print("group {}".format(m.group(1)))
            res = [clinsig,"{}:{}:{}".format(GRCh37Chromosome, GRCh37Location, m.group(1)),
                  "{}:{}:{}".format(GRCh37Chromosome, GRCh38Location, m.group(1))]
            print(res)
            if clinsig in outputsigs:
                outputDict[outputsigs[clinsig]].append(res[1:3])


    #print(nm, clinsig, GRCh37Location)

    
#print(json.dumps(outputDict, indent=3)) 
file1.close()

outputfile = open('/Users/forei/Downloads/clinvar_BRCA2_full.txt', 'w')

for vtype, vList in outputDict.items():
    for v in vList:
        outputfile.write("{}\t{}\t{}\n".format(vtype, v[0], v[1]))
outputfile.close()
```
